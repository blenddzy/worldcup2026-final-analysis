# 03 - Visualizations
Generating plots for the World Cup 2026 final

```python
import sys
sys.path.insert(0, '..')
import matplotlib.pyplot as plt
import pandas as pd
import json

from src.visualize import plot_goal_timeline, plot_tournament_path, plot_group_standings
from src.analyze import route_to_final, extract_goals_by_team
from src.clean import load_openfootball, standings_to_df
```

```python
matches_df = load_openfootball('../data/raw/openfootball_2026.json')

with open('../data/raw/openfootball_2026.json') as f:
    raw = json.load(f)
final_match = [m for m in raw['matches'] if m['round'] == 'Final'][0]
```

```python
goals_df = extract_goals_by_team(final_match)
plot_goal_timeline(goals_df, '../reports/figures/goal_timeline.png')
```

```python
spain_route = route_to_final(matches_df, 'Spain')
arg_route = route_to_final(matches_df, 'Argentina')
plot_tournament_path(spain_route, arg_route, '../reports/figures/route_to_final.png')
```

```python
standings_df = standings_to_df(raw.get('standings', {}))
if not standings_df.empty:
    plot_group_standings(standings_df, '../reports/figures/group_standings.png')
```
